```python
df[['Service_year_of_vehicle','Defect_of_vehicle','Type_of_vehicle','Accident_severity']]=df[['Service_year_of_vehicle','Defect_of_vehicle','Type_of_vehicle','Accident_severity']].astype(str)
```

```python
df[['Service_year_of_vehicle','Defect_of_vehicle','Type_of_vehicle','Accident_severity']]=df[['Service_year_of_vehicle','Defect_of_vehicle','Type_of_vehicle','Accident_severity']].fillna(df[['Service_year_of_vehicle','Defect_of_vehicle','Type_of_vehicle','Accident_severity']].mode())
```

```python
df[['Fitness_of_casuality','Work_of_casuality','Driving_experience','Types_of_Junction']]=df[['Fitness_of_casuality','Work_of_casuality','Driving_experience','Types_of_Junction']].fillna(df[['Fitness_of_casuality','Work_of_casuality','Driving_experience','Types_of_Junction']].astype(str))
```

```python
df[['Fitness_of_casuality','Work_of_casuality','Driving_experience','Types_of_Junction']]=df[['Fitness_of_casuality','Work_of_casuality','Driving_experience','Types_of_Junction']].fillna(df[['Fitness_of_casuality','Work_of_casuality','Driving_experience','Types_of_Junction']].mode())
```

```python
df[['Casualty_severity']]=df[['Casualty_severity']].fillna(df[['Casualty_severity']].mode())
```
