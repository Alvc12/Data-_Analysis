## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Task (2) Dataset.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.sample(10)
```

```python
df.isnull().sum()
```

```python
df['TravelTimeHistoric'].fillna(df['TravelTimeHistoric'].mean(), inplace=True)
df['TravelTimeLive'].fillna(df['TravelTimeLive'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

```python
df.info()
```

### Identify and analyze patterns in the data.

```python
df.describe()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['TrafficIndexLive'], bins=30, kde=True, color='blue')
plt.title('Distribution of Traffic Index')
plt.xlabel('Traffic Index')
plt.ylabel('Frequency')
plt.show()
```

```python
df.plot.scatter(x='JamsCount', y='JamsLength')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


## 2. Data Visualization

```python
plt.figure(figsize=(8, 8))
df['City'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.ylabel('')
plt.show()
```

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df['City'], y=df['JamsCount'], palette='pastel')
plt.title('Traffic Jams Count by City')
plt.xlabel('City')
plt.ylabel('Number of Traffic Jams')
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='JamsDelay', y='TravelTimeLive', color='pink')
plt.title('Travel Time Live vs. Jams Delay')
plt.xlabel('Jams Delay (minutes)')
plt.ylabel('Travel Time Live (minutes)')
plt.show()
```
