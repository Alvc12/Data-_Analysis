```python
df['City']=df['City'].astype(str)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.lineplot(data=df, x='TrafficIndexLive',y='TrafficIndexWeekAgo')
# here we can see the changes in traffic index live and the traffic index week age
```

```python
numerical_variable=df[['TrafficIndexLive',	'JamsCount',	'JamsDelay'	,'JamsLength'	,'TrafficIndexWeekAgo',	'TravelTimeHistoric',	'TravelTimeLive']]
corrrelation_mat=numerical_variable.corr()
sns.heatmap(corrrelation_mat,annot=True)

#I've used heatmap to show the correlation between numerical varaibles
```

```python
sns.scatterplot(x=df['JamsDelay'],y=df['JamsLength'])
#where i've used scatterplot to show the relationships between the Jams Length and Jams Delay
```

```python
import plotly.express as px
fig=px.pie(names=df['City'],values=df['JamsCount'])
fig.show()
```
